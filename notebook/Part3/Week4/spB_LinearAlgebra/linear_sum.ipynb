<h2> 선형대수 발표 요약</h2>

<h3><요약 1> 선형종속 = deteminant 0 = 특이행렬 singular matrix = 역행렬 없음</h3>

이미지로 생각하기 쉽게하기 위해 정방행렬로 적어보겠습니다.

<pre>
A = [1 2 3
     4 5 6
     7 8 9]
     
  = [x<sub>1</sub>, x<sub>2</sub>, x<sub>3</sub>]
</pre>

위 A 라는 행렬은 x<sub>1</sub>, x<sub>2</sub>, x<sub>3</sub> 라는 세개의 벡터로 이루어진 것으로 볼 수 있습니다.

(행렬을 벡터들의 모임으로 보면 좀 편한것 같습니다.)

---

<h3><요약 2>'선형종속'을 판단하는 아래 식</h3>

(k 는 스칼라값, 단 하나이상의 0이 아닌 실수가 존재)

> **k<sub>1</sub>x<sub>1</sub> + k<sub>2</sub>x<sub>2</sub> + k<sub>3</sub>x<sub>3</sub> = 0**

다른식으로 표현하면, (k<sub>1</sub> 이 0이 아닌 것이라 하자) 아래와 같습니다. 

> k<sub>2</sub>x<sub>2</sub> + k<sub>3</sub>x<sub>3</sub> = -k<sub>1</sub>x<sub>1</sub>

즉 바꿔설명하자면, 

선형 종속이란, 행렬을 이루는 **'어떤 벡터를 나머지 벡터들의 합으로 표현할 수 있느냐, 그것이 하나 이상 있느냐'**를 말합니다.

반대는 어떤벡터가 다른벡터들로 표현할 수 없는 요소를 가지고 있느냐입니다.

위 식을 판별하기 위해서는 행렬식determinant 를 계산해서 0 이 나오는지 확인하면 됩니다.

---

<pre>
[1 2 4
 1 3 5
 1 4 6]
</pre>

2x<sub>1</sub> + x<sub>2</sub> = x<sub>3</sub>

x<sub>3</sub> 벡터는 나머지 벡터의 조합으로 표현이 가능합니다. = 선형종속

---

이제 이것을 또다시 다르게 표현하면,

(보통) 3차원 벡터 3개가 모이면 3차원 공간을 표현할 수 있는데,

(x<sub>1</sub>, x<sub>2</sub> 둘은 서로 독립이라고 치자)

위 경우는 **x<sub>1</sub>, x<sub>2</sub> 두 벡터가 이루는 평면에 x<sub>3</sub> 벡터가 놓여있는 상황**입니다.

즉, 3 차원 **벡터 3개가 모였는데 조합을 해도 온전한 3차원을 나타낼 수 없는 상태**입니다.

독립적이지 못한 벡터가 존재하는 상태.

이때가 '선형종속' = deteminant 0 = '특이행렬 singular matrix' = 역행렬없음

---

만약 벡터 중 하나라도 0(영)벡터가 들어가 있으면 반드시 선형종속인데,

이유는 k<sub>2</sub>x<sub>2</sub> + k<sub>3</sub>x<sub>3</sub> = 0 (k<sub>1</sub>x<sub>1</sub>)

k<sub>1</sub> != 0 이고 나머지 k 값이 모두 0이면 위 식을 만족하기 때문입니다. 

따라서 선형종속이고, det(A) = 0, 이 되고 역행렬이 존재하지 않습니다.

<h3><요약 3>**계수 rank 라는 것은 선형독립인 벡터들의 갯수**입니다.</h3>

다른 벡터들의 조합으로 만들어 낼 수 없는 벡터의 갯수입니다.

쉽게 말하면, 3차원 좌표의 x,y,z 벡터들이 선형독립인 경우이고, 

아무리 x,y 축 벡터들을 조합해도 z 벡터를 만들어 낼 수 없습니다.

x,y,z 각각 벡터은 다른 벡터들의 조합으로 나타낼 수 없고, 이때 rank = 3 입니다. 

만약 3 벡터들이 같은 평면에 존재하면 rank = 2,

모두 같은 직선에 존재하면 rank = 1 입니다.

그래서, 3차 정방행렬 A 가 가질 수 있는 최대 계수 rank 는 3개입니다.

해당 행렬이 가질 수 있는 최대 rank 를 가지지 못하면 선형종속이란 뜻이고, 특이행렬입니다.

<h3>그런데 선형종속이 되면 왜 역행렬이 존재하지 않을까?</h3>

3차 정방행렬로 계속 생각해보겠습니다.

위에서 예를 든 선형종속의 경우, 3개의 벡터는 실제로 조합을 통해 3차원 공간이 아닌 평면에 나타내고 있습니다.

그런데 역행렬이란 무엇일까요?

> A * ? = I 

항등행렬을 만드는 행렬입니다.

그런데 항등행렬 I 는 각 차원 x,y,z 에 대해서 축방향(온전한 3차원)을 나타내는 행렬입니다.

<pre>
I = [1 0 0
     0 1 0
     0 0 1]
  = [i<sub>1</sub>, i<sub>2</sub>, i<sub>3</sub>]
</pre>

요약하자면, 

**선형종속으로 인해 온전한 3차원을 나타내지 못하는 A 행렬은 어떤 행렬을 곱하던 온전한 3차원을 표현하는 행렬로 바꿀 수 없습니다.**

(간단히 증명해보자면, determinant 의 속성에 보면, 

|AB| = |A||B|

행렬 곱한 전체의 det 값은 각각의 det 값의 곱과 같다고 했는데요.

A 가 특이행렬이라 |A|값이 0 이면, 위 속성에 따라 어떤 행렬을 곱해도 det = 0 입니다.

<h3><요약 4> 즉, 단순 행렬 곱을 해봐자, 한 번 특이행렬은 영원한(?)특이행렬입니다.</h3>

그래서 특이행렬에 무엇을 곱하던, **비특이행렬인 항등행렬**을 만들 수 없는 것입니다.

항등행렬 뿐 아니라 일반 비특이행렬도 만들 수 없습니다.

비특이행렬로 돌아올 수 없는 강을 건넜습니다. ㅠㅠ)

---

(실제로 항등행렬을 만드는 x 를 구할 수 없습니다.
 
왜냐하면, 선형종속에 의해 한벡터가 다른벡터들의 합으로 표현이 가능한 행렬 A 에 대해, A * X 변수로 넣고 전개하면, 

(간단하게 2차 행렬 A 일때)

1x<sub>1</sub> + 2x<sub>2</sub> 와 2x<sub>1</sub> + 4x<sub>2</sub> 와 같은 중복 꼴의 식이 나오는데, 

> 1x<sub>1</sub> + 2x<sub>2</sub> = 1

> 2x<sub>1</sub> + 4x<sub>2</sub> = 0 

서로 다른 값인 항등행렬의 요소 1, 0 이 나와야되는 식이 발생하고, 해를 구할 수 없는 방정식이 됩니다.)

<h3> singular matrix? </h3>

선형종속에 의해 역행렬이 없는 행렬을 **'특이행렬 singular matrix'** 라고 합니다.

특이행렬은 特異 라는 뜻의 '특이한, 특별한' 이라는 뜻을 가지고 있고, 

singular 단어의 뜻에 '단수' 외에 '특이한' 이라는 뜻을 가지고 있습니다.

왜 특이하다라고 했을까?

2차원 벡터 두 개를 모았는데 같은 직선에 놓일 경우는 아마 휘귀하겠죠;

3차원 벡터 3 개를 모았는데 한 평면이나 한 직선에 놓인 경우는 특이하겠죠;;;

이처럼 선형종속을 만드는 경우는 경우의 수로 치면 특이한 케이스가 되는 듯 합니다.

<h3> 고유값 eigen value, 고유벡터 eigen vector </h3>

<h3><요약 6>**정방행렬 A 에 대해 (크기빼고) 방향은 변하지 않는 벡터를 찾는 것이다.**</h3>

나중에 좌표축(base)를 바꿔서 문제를 변형할때 사용하는 듯 한데, 아직은 목적이나 사용예제를 잘 모르겠습니다. -_-;;

(A 는 행렬, x 벡터, L 스칼라값)

> Ax = Lx

위 식을 만족하는 고유값 L, 고유벡터 x 쌍의 갯수는 0개 ~ A행렬차원수 이다. (설명을 쉽게하기 위해 정방행렬만 따져보겠음)

> (A - LI) x = 0 (영행렬)

(고유벡터가 있다면) A-LI 행렬은 특이행렬이 되어 역행렬이 없어야 한다. 

|A-LI| = 0 식을 전개해서 L 에 대한 다항식 풀이를 통해 고유값들을 구하고,

구한 고유값을 넣고 고유벡터를 구하면 된다. 

sc.eig(A) 를 사용하면, 쉽게 고유값, 고유벡터를 구해준다.
