## Cross-correlation

- 두 신호(함수, 벡터, 이미지)의 유사도를 측정
- 상관계수처럼 값으로 표현되는 것이 아니라 τ(time-lag, 시간지연)에 대한 함수로 표현
- ![](http://upload.wikimedia.org/wikipedia/commons/6/6a/Convolution_of_box_signal_with_itself2.gif)
  - 함수 g를 얼마나 이동(slide, shift) 시켰을 때 두 함수가 가장 유사해지는 지를 관찰
  - 이 신호에서 주어지 패턴이 언제(어디에) 나타나는 지
  - 이 신호에서 주어진 패턴이 반복적으로 나타나는 지

![](imgs/correlation_definition.png)
- *는 복소켤레를 의미
  - align 되었을 때 값이 증가하도록 하기 위해
- 이산함수에서는 적분대신 시그마로 정의됨
- [5, 4, 1, 3, 2]와 [1, 2, 4, 0]의 Cross-correlation 예시
![](imgs/small_digital_version_of_correlation.png)
- 기본정의대로 계산하려면 시간복잡도가 O(n²)이됨.
- 계산량을 줄이기 위해 푸리에 변환을 이용
![](imgs/correlation_fourier_transformed_final.png)
  - F, G는 각각 f, g의 푸리에 변환
  - 결과값을 역 푸리에 변환하여 계산
  - 계산량이 3N log2 N + N로 줄어듬

## Correlation of DNA Strings

http://www.ncbi.nlm.nih.gov/pmc/articles/PMC2291754/

전통적인 sequence alignment algorithms들보다
- deeply hidden sequence homology도 잘 찾고
- adjustable parameters necessary for gap penalties도 필요없고
- intrasequence rearrangements도 잘 잡아낼 수 있으며
- FFT를 이용해 구현하면 속도마저 빠르다!

### 구현

```python
from pysrc import foura
```

```python
def nucleotide2num(nucleotide):
    nucleotides = 'acgt'
    nums = [1j, -1, 1, -1j]  # complementary bases having opposite signs
    return nums[nucleotides.index(nucleotide)]
nucleotide2num = vectorize(nucleotide2num, otypes=[complex_])


full_sequence = list(foura.GetData('genbank/NC_000962.3.gb')[1])
pyrG_sequence = list(foura.GetData('genbank/885048.gb')[1])

print len(full_sequence)

k = 1000
full_sequence = full_sequence[1900*k:2000*k]
len_full_sequence = len(full_sequence)
len_pyrG_sequence = len(pyrG_sequence)

print len_full_sequence
print len_pyrG_sequence

converted_full_sequence = nucleotide2num(full_sequence)
converted_pyrG_sequence = nucleotide2num(pyrG_sequence)

# akando can calculate only vectors with the same size
converted_pyrG_sequence = concatenate((converted_pyrG_sequence, zeros(len_full_sequence - len_pyrG_sequence)))

corr = akando.Correlate(converted_full_sequence, converted_pyrG_sequence)
plot(corr.real)
```

## Two-Dimensional Correlations

### Normalized Cross-Correlation

- 이미지간 밝기, 노출 등의 차이를 상쇄하기 위해 Normalize 후 처리하게 된다.
- Normalize 된 값은 곧 피어슨 상관계수가 된다.

### 구현

- source image ![](imgs/source_image.jpg)
- template image ![](imgs/template_image.jpg)

```python
from scipy import signal
from PIL import Image

def normalize(data):
    normalized = (data - data.mean()) / data.std()
    return normalized

source_image = Image.open('imgs/source_image.jpg')
template_image = Image.open('imgs/template_image.jpg')

source_image_grayed = source_image.convert('L')
template_image_grayed = template_image.convert('L')

source_data = akando.i2a(source_image_grayed)
template_data = akando.i2a(template_image_grayed)

normalized_source_data = normalize(source_data)
normalized_template_data = normalize(template_data)

corr = signal.correlate2d(normalized_source_data, normalized_template_data, mode='same')
```
