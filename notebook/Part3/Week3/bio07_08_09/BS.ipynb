Step 6: Convert the path to an alignment

– Vertical move   = gap in the top sequence
– Horizontal move = gap in the left sequence
– Diagonal move   = alignment of nucleotides from each sequence

```python
import numpy as np

#seq1 = "ACACT"
#seq2 = "ACT"
seq1  = "AAACTTACACGAATCGCTTTCTGGAGGAGGAGGGGACCCGCTGCGCG"
seq2  = "AACCTTACACGAATCGCTTTCTGAGGAGGGCACCCGCTGCGCG"
#seq1  = "AAACTTACACGAATCGCTTTCTGGAGGAGGAGGGGACCCGCTGCGCG"
#seq2  = "TCGCTTTCTGAGGA"
#seq1   = "ATTCGATCC"
#seq2   = "ACGAT"
match       = 1
nonmatch    = 0
gap_penalty = -1
bLocal       = 1 # 1 : local alignment, 0 : global alignment

#ATGCATGC
#AT--ATGC

table = np.zeros((len(seq2)+1,len(seq1)+1)) # seq2 goes to row
#print table[0]
#print np.arange(-1,-1*(len(seq2)+1)-1,-1)


# initial gap sequence
if bLocal == 0:
    table[0]   = np.arange(0,-1*(len(seq1)+1),gap_penalty)
    table[:,0] = np.arange(0,-1*(len(seq2)+1),gap_penalty)
else:
    pass






for nr in range(len(table)):
    for nc in range(len(table[nr])):
        if nr > 0 and nc > 0:
                pass
        else: continue
        #print(nr,nc)
        if seq2[nr-1] == seq1[nc-1]:
            match_value = match
        else: match_value = nonmatch
        vm = table[nr-1,nc] + gap_penalty # vertical move
        hm = table[nr,nc-1] + gap_penalty # horizontal move
        dm = table[nr-1,nc-1] + match_value # diagonal move
        selected_value = max(vm,hm,dm)
        if bLocal == 1:
            if selected_value < 0:
                selected_value = 0
            else:
                pass
        else:
            pass
        table[nr,nc] = selected_value

print table       
        
Outseq1 = ''
Outaln  = ''
Outseq2 = ''

maxnc = len(table[nr])
maxnr = len(table)
#print(len(seq1))
#print(len(seq2))

## rnc : reverse number of column, start point of searching alignment path
if bLocal == 1:
    rnc = table[-1].argmax()-1
else:
    rnc = maxnc-1  # assume last base is matched
rnr = maxnr-1  

while rnc > 0 and rnr > 0:
      
    if rnc-1-1 < 0 or rnr-1-1 < 0:
        break
    window = np.array([table[rnr-1,rnc-1],table[rnr-1,rnc],table[rnr,rnc-1]])
    if window.argmax() == 0:
        Outseq1 += seq1[rnc-1-1]
        Outseq2 += seq2[rnr-1-1]
        Outaln  += '|'
        rnc     -= 1
        rnr     -= 1
    elif window.argmax() == 1:
        Outseq1 += '-'
        Outseq2 += seq2[rnr-1-1]
        Outaln  += ' '
        rnr     -= 1
    else:
        Outseq1 += seq1[rnc-1-1]
        Outseq2 += '-'
        Outaln  += ' '
        rnc     -= 1
print "input sequence 1 :",seq1
print "input sequence 2 :",seq2
print "Alignment result"
print Outseq1[::-1] + seq1[-1]
print Outaln[::-1] + '|'
print Outseq2[::-1] + seq2[-1]
            
            
        
        
```

sequence에서 word를 추출한다

```python
import numpy as np

def wordextract(seq,word_size):
    words = []
    seq_len = len(seq)
    for i in range(int(seq_len) - word_size +1):
        words.append(seq[i:i+word_size])
    return list(words)
seq = 'ATGACAACTGATGCATGA'
print wordextract(seq,4)
```

sequence에서 특정 word의 위치를 가져온다.

```python
def findoccur(seq,word):
    count = seq.count(word)
    locs = []
    k = 0
    for i in range(count):
        k = seq.find(word,k)
        locs.append(k)
        k += 1
    return(locs)
print(findoccur(seq,'AT'))
```

```python
def histogram(upper,distances):
    Nbins = upper + 1
    hist = zeros(Nbins,int)
    for i in range(len(distances)):
        k = distances[i]
        if 0<k<Nbins: hist[k] += 1
    return(hist)
print histogram(40,[10, 14,  4, 22, 12,  8, 55, 45, 41,33])
        
```
