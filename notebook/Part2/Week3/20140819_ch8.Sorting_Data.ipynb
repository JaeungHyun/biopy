#### sort() vs. sorted()
* lists vs. any iterable data structure
* modifies the list in-place vs. builds a new list

```python
# sort()
data = [1, 5, 7, 8, 9, 2, 3, 6, 6, 10]
data.sort()
data
```

```python
# sort in descending order using sort() function
data = [1, 5, 7, 8, 9, 2, 3, 6, 6, 10]
data.sort()
data.reverse()
data
```

```python
# when you sort a list of lists, the sorting is carried out based on the first element of each list
list_of_lists = [[1,2], [4,2], [9,1], [2,7], [2,5]]
list_of_lists.sort()
list_of_lists
```

```python
# Altenatively, you may pass an cmp() function as an argument to the sort() method
# cmp(a,b) returns a negative value if a<b, zero if a=b, or a positive value if a>b
# The cmp() function can be customized
def my_cmp(a,b):
    if a > b : return -1
    if a == b : return 0
    if a < b : return 1
L = [1, 2, 3, 4, 5, 6, 8, 8, 9, 9, 30]
L.sort(my_cmp)
L
```

```python
# sorted()
data = [1, 5, 7, 8, 9, 2, 3, 6, 6, 10]
newdata = sorted(data)
newdata
```

#### Sorting Data Structure (Tuples, Dictionaries)

```python
# Sort a tuple using sort()
data = (1, 4, 5, 3, 8, 9, 2, 8, 9, 30)
list_data = list(data)
list_data.sort()
new_tup = tuple(list_data)
print new_tup
```

```python
# Sort a tuple using sorted()
data = (1, 4, 5, 3, 8, 9, 2, 8, 9, 30)
new_data = tuple(sorted(data))
print new_tup
```

```python
# Sort a dictionary using sort()
data = {1:'a', 2:'b', 4:'d', 3:'c', 5:'t', 6:'m', 36:'z'}
keys = list(data)
keys
```

```python
keys.sort()
for key in keys:
    print key, data[key]
```

```python
# Sort a dictionary using sorted()
data = {1:'a', 2:'b', 4:'d', 3:'c', 5:'t', 6:'m', 36:'z'}
for key in sorted(data):
    print key, data[key]
```

#### Sorting Strings by Their Length
* You can use the sorted() function with a lambda function as a custom parameter instead of itemgetter

```python
data = ['ACCTGGCCA', 'ACTG', 'TACGGCAGGAGACG', 'TTGGATC']
from operator import itemgetter
itemgetter_data = sorted(data, key = itemgetter(1))
itemgetter_data
```

```python
data = ['ACCTGGCCA', 'ACTG', 'TACGGCAGGAGACG', 'TTGGATC']
bylength_data = sorted(data, key = lambda x: len(x))
bylength_data
```
