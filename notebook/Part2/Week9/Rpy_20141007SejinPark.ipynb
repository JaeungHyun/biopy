## Calculating	mean,	standard	deviation,	z-score	and	p-value	 of	a	set	of	numbers

### in Python:

```python
import rpy2.robjects as ro
r = ro.r
```

```python
f = r("read.table('RandomDistribution.tsv',sep='\t')")
```

```python
m = r.mean(f[2], trim = 0, na_rm= 'FALSE')
```

```python
sdev= r.sd(f[2], na_rm= 'FALSE')
```

```python
value = 0.01844
```

```python
z = (m[0] - value)/sdev[0]
```

```python
x = r.abs(z)
```

```python
p = r.pnorm(-x[0])
```

```python
p[0]
```

## Plot	random	numbers	to	a	file	

```python
import rpy2.robjects as ro
from rpy2.robjects.packages import importr
r = ro.r
grdevices= importr('grDevices')
grdevices.png(file="RandomPlot.png", width=512, height=512)
r.plot(r.rnorm(100), ylab= "random")
grdevices.dev_off()
```

```python
from IPython.display import Image
Image(filename='RandomPlot.png')
```

## Plot	data	data	from	a	file

```python
import rpy2.robjects as ro
from rpy2.robjects.packages import importr
r = ro.r
f = r("read.table('RandomDistribution.tsv',sep='\t')")
grdevices= importr('grDevices')
grdevices.png(file="Plot.png", width=512, height=512)
r.plot(f[1],f[2], xlab= "x", ylab= "y")
grdevices.dev_off()
```

```python
from IPython.display import Image
Image(filename='Plot.png')
```

## Plot	data	data	from	a	file	

```python
grdevices.png(file="Histogram.png", width=512, height=512)
r.hist(f[4], xlab='x', main = 'Distribution of values')
grdevices.dev_off()
```
