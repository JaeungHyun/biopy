<h1>Ch.20 Retrieving Data from Web Resources</h1>

use Biopython to <FONT color="navy">1) access online NCBI databases; such as PubMed, GenBank, and Expasy</FONT>, and <FONT color="navy">2) retieve and parse their contents.</FONT>

In this chapter : find, retrieve and parse publications about 'PyCogent' in NCBI>PubMed

<b>Bio.Entrez</b> : access and download NCBI database records.
 - Entrez.esearch(db, term, **keywds) : search in NCBI databases -> Entrez.read
 - Entrez.egquery(term) : count the number of matches in each of the Entrez databases.
 - Entrez.efetch(db, **keywds) : download the records
 - Entrez.einfo(**db) : 
     - without argument : get a dictionary ['DbList']:[databases in Entrez]
     - with argument : get a dictionary ['DbInfo']:[information about that database]
 
 
<b>Bio.Medline</b> : parse the records.
 - Medline.Record : dictionaries of the results

```python
from Bio import Entrez
from Bio import Medline
Entrez.email = "jadeport0320@gmail.com"
keyword = "PyCogent"
```

<b>Entrez.esearch(<FONT color="navy">db</FONT>, <FONT color="navy">term</FONT>, <FONT color="navy">**keywds</FONT>)</b>

 - <FONT color="navy"><b>db</b></FONT> : http://www.ncbi.nlm.nih.gov/books/NBK25497/#chapter2.chapter2_table1<br>default = "pubmed"</br>
 - <FONT color="navy"><b>term</b></FONT> : A[field] (AND[OR] B[field]) 
 http://www.ncbi.nlm.nih.gov/books/NBK3827/#pubmedhelp.Search_Field_Descrip <FONT color="red"># example 2</FONT>
 - <FONT color="navy"><b>**keywds</b></FONT> : retmax(maximum retrieved), datetype(mdat, pdat, edat), mindate, maxdate("YYYY/MM/DD")


```python
handle = Entrez.esearch(db="pubmed", term=keyword)
record = Entrez.read(handle)
pmids = record['IdList']
print pmids
```

<b># record.keys()</b> : ['Count', 'RetMax', 'IdList', 'TranslationStack', 'RetStart', 'QueryTranslation']

```python
handle = Entrez.esearch(db="pubmed", term="Python",\
                        datetype="pdat",mindate="2011/08/01", maxdate="2014/9/17")
## mdat = modification date, pdat = publication date, edat = Entrez date
record = Entrez.read(handle)
print record['Count']
```

<b>Entrez.efetch(<FONT color="navy">db</FONT>, <FONT color="navy">**keywds</FONT>)</b>

 - <FONT color="navy"><b>db</b></FONT> : http://www.ncbi.nlm.nih.gov/books/NBK25497/#chapter2.chapter2_table1<br>default = "pubmed"</br>

 - <FONT color="navy"><b>**keywds</b></FONT> : retmode, rettype, retstart, retmax
 <br>Valid values of &retmode and &rettype for EFetch :
 
 http://www.ncbi.nlm.nih.gov/books/NBK25499/table/chapter4.T._valid_values_of__retmode_and/?report=objectonly</br>


<b># record[key] : record[value]</b>

 - 'TI' : Title
 - 'PMID'
 - 'PG' : pages
 - 'AB' : Abstract
 - 'AU' : Authors
 
 and so on..

```python
handle = Entrez.efetch(db="pubmed", id=pmids, rettype="medline",retmode="text")
medline_records = Medline.parse(handle)
records = list(medline_records)
print 'keys : ', records[0].keys()
n = 1
for record in records:
    if keyword in record["TI"]:
        print n, ')', record["TI"]
        n += 1
## http://www.ncbi.nlm.nih.gov/pubmed/22479120?report=medline&format=text
```

##### Example 2 Searching PubMed with more than one term, combining keywords with AND/OR

```python
from Bio import Entrez

handle = Entrez.esearch(db="pubmed", term="PyCogent AND RNA")
record = Entrez.read(handle)
print record['IdList']

handle = Entrez.esearch(db="pubmed", term="PyCogent OR RNA")
record = Entrez.read(handle)
print record['Count']

handle = Entrez.esearch(db="pubmed", term="PyCogent AND 2008[Year]")
record = Entrez.read(handle)
print record['IdList']

handle = Entrez.esearch(db="pubmed", term=\
                        "C. elegans[Organism] AND 2008[Year] AND Mapk[Gene]")
record = Entrez.read(handle)
print record['Count']
```

```python
handle = Entrez.esearch(db="pubmed", \
                        term = "PyCogent OR RNA", retmax = "3")
record = Entrez.read(handle)
print record['IdList']
```

##### Example 3 Retrieving and parsing Nucleotide database entries in GenBank Format

```python
from Bio import Entrez

# search sequences by a combination of keywords
handle = Entrez.esearch(db="nucleotide", \
                        term="Homo sapiens AND mRNA AND MapK")
records = Entrez.read(handle)
print records['Count']

top3_records = records['IdList'][0:3]
print top3_records

# retrieve the sequences by their GI numbers
gi_list = ','.join(top3_records)
print gi_list
handle = Entrez.efetch(db="nucleotide",\
                       id=gi_list, rettype="gb", retmode="xml")
records = Entrez.read(handle)
print len(records)
print records[0].keys()
print records[0]['GBSeq_organism']
```

```python
# remode="text" can be used for a single gi
handle = Entrez.efetch(db = "nucleotide", \
                       id="186972394", rettype="gb", retmode="text")
record = handle.read()

print record
```

##### Example 4 Searching for NCBI Protein database entries by keywords

```python
from Bio import Entrez

# search IDs of protein sequences by keywords
handle = Entrez.esearch(db="protein",term="Human AND cancer AND p21")
records = Entrez.read(handle)
print records['Count']
id_list = records['IdList'][0:3]

# retrieve sequences
id_list = ",".join(id_list)
print id_list

handle = Entrez.efetch(db="protein",id=id_list, rettype="fasta", retmode="xml")
records = Entrez.read(handle)
rec = list(records)
print rec[0].keys()
print rec[0]['TSeq_defline']
```

##### Example 5 Retrieving SwissProt database entries and writing them to a file in FASTA format

```python
from Bio import ExPASy
from Bio import SeqIO

handle = ExPASy.get_sprot_raw("P04637") # can be read using SeqIO.read
seq_record = SeqIO.read(handle, "swiss")
out = open('myfile.fasta','w')
# SeqIO.write(id, Seq, description)
fasta = SeqIO.write(seq_record, out, "fasta")
out.close()
```

```python
ac_list = ['P04637', 'P0CQ42', 'Q13671']
records = []
for ac in ac_list:
    handle = ExPASy.get_sprot_raw(ac)
    record = SeqIO.read(handle, "swiss")
    records.append(record)
out = open('myfile.fasta','w')
for rec in records:
    fasta = SeqIO.write(rec, out, "fasta")
out.close()
```

HTMLParser.feed() # which parses HTML string passed as an argument and calls custom method to take care of individual tags.

```python
from HTMLParser import HTMLParser
import urllib

class MyHTMLParser(HTMLParser):
    def handle_starttag(self, tag, attrs):
        self.start_tag = tag
        print "Start tag:", self.start_tag
    
    def handle_endtag(self, tag):
        self.end_tag = tag
        print "End tag:", self.end_tag
    
    def handle_data(self, data):
        self.data = data.strip()
        if self.data:
            print "Data :", self.data

parser = MyHTMLParser()
url = 'http://www.ncbi.nlm.nih.gov/pubmed/21998472'
page = urllib.urlopen(url)
data = page.read()
parser.feed(data)
# handle_starttag(), handle_endtag(), and handle_data() called
```
