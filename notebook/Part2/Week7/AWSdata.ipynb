```python
tbody = soup.find_all('tbody')[1]
print tbody.text
```

```python
hr_today = pd.DataFrame([i.split('\n')[1:15] for i in tbody.text.split('\n\n\n')[1:]],\
                   index=[i.split('\n')[0] for i in tbody.text.split('\n\n\n')[1:]],\
                   columns=tbody.text.split('\n\n\n')[0].split('\n')[3:], dtype=np.float64)
hr_today.index.name = 'Observation_time'
hr_today.columns.name = 'Variables'

print hr_today
```

```python
hr_today.columns = ['avgTemp', 'maxTemp', 'minTemp', 'RH', 'SR', 'SM', 'WD', 'LW',\
                    'soilTemp1', 'soilTemp2', 'avgMXWS', 'Rain', 'maxMXWS', 'Bat']

print hr_today.ix[hr_today.index[0]]['RH'] # 가장 최근의 상대습도 값 (특정 열, 행의 데이터)
print hr_today['RH'].mean() # 오늘의 상대습도 평균 값 (세로 열의 평균)
print hr_today['LW'].sum() # 오늘 엽면습윤 시간의 총합 (세로 열의 총합)
print hr_today[hr_today.columns[7]].sum() # 동일한 방법
print hr_today.ix[hr_today.index[0]].sum() # 가장 최근의 모든 데이터의 총합 (가로 열의 총합)
```
