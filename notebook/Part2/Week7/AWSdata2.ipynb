```python
# retrieve daily data of August
rawurl = 'http://weather.stacorp.co.kr/weather/weather1_2.asp?excel=0&NU_SITE_CD=34&NU_STATION_CD=1&txt_date=2014-08-01&txt_edate=2014-08-31'
url1 = br.open(rawurl)

read2 = url1.read()
soup2 = BeautifulSoup(read2)
tbody = soup2.find_all('tbody')[1]
print soup2.find_all('tbody')[1].text
```

```python
dataAug = pd.DataFrame([i.split('\n')[1:15] for i in tbody.text.split('\n\n\n')[1:]],\
                   index=[i.split('\n')[0] for i in tbody.text.split('\n\n\n')[1:]],\
                   columns=tbody.text.split('\n\n\n')[0].split('\n')[3:], dtype=np.float64)
dataAug.index.name = 'Observation_time'
dataAug.columns.name = 'Variables'

print dataAug
```

```python
# modification of index format with datetime module
Aug1st = datetime.date(2014,8,1)
dataAug.index = [(Aug1st + datetime.timedelta(days=i)).strftime('%a %d %b') \
                 for i in range(0,31)]
dataAug.columns = ['avgTemp', 'maxTemp', 'minTemp', 'RH', 'SR', 'SM', 'WD', 'LW',\
                    'soilTemp1', 'soilTemp2', 'avgMXWS', 'Rain', 'maxMXWS', 'Bat']
print dataAug
```
